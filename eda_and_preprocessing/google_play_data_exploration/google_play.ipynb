# outlier treatment
#we can observe from above three graps that rating column has outlier


```python
plt.figure(figsize=(10,10))
sns.boxplot(df['Rating'])
plt.show()
```

```python
q1=np.quantile(df['Rating'],0.25)
q3=np.quantile(df['Rating'],0.75)
iqr=q3-q1
```

```python
#finding value of upper whisker and ower whisker
upper_whisker=q3+3*iqr
lower_whisker=q1-3*iqr
print(upper_whisker,lower_whisker)
```

```python
df.shape
```

```python
df=df[df['Rating']<upper_whisker]
df.shape
```

```python
#so there was one outlier only
#lets check boxplot again
plt.figure(figsize=(10,10))
sns.boxplot(df['Rating'])
plt.show()
```

```python
#lets see how many outliers below lower whisker
new_data=df[df['Rating']>lower_whisker]
new_data.shape
```
