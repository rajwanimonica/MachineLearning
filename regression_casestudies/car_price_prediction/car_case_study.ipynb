linear regression aumption

```python
#features we are using for model training must have co relation with target data i.e price
#data must be homoscedastic
#if scatter plot of feature vs target is scatter i.e if feature and target have  low co relation we should not  use them for model training
#features we are using should not have multicolinearity or interdependencies
#if feature and target have high oulier not suggested to use linear regression 
```

```python
#price has high co-relation with carwidth, curbweight,enginesize,horsepower and citympg and highway mpg
# but curbweight, horsepower,citympg and highway mpg have interdependency with enginesize.and among all the above features enginesize has highest corelationship with price
#scatterplot of enginesize vs price is homoscedastic
#so the features we are going to use could be carwidth and enginesize
```

```python
new_df=df[['price','enginesize','carwidth']]
new_df
```

```python
from sklearn.model_selection import train_test_split as tts
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import MinMaxScaler
# scale=MinMaxScaler()
# scaled_df=scale.fit_transform()
xtrain, xtest, ytrain,ytest=tts(new_df[['enginesize','carwidth']],df['price'], test_size=0.30)
lr=LinearRegression()
lr.fit(xtrain,ytrain)
lr.score(xtrain,ytrain)
```

```python
yhat=lr.predict(xtest)
yhat
```

```python
ytest
```

```python
from sklearn.metrics import r2_score
r2_score(ytest,yhat)
```
