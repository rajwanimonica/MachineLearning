```python
#medv is our target
#and medv is our target i.e MEDV: Median value of owner-occupied homes in $1000s
# By looking at the correlation matrix we can see that rm has a strong positive correlation with medv(0.7)
#lstat has a high negative correlation with medv(-0.74).
# The features rad, tax have a correlation of 0.91. These feature pairs are strongly correlated to each other. but rad is not corelated with medv.so we should not take them together or training model
#dis and age which have a correlation of -0.75.so we should not take them them together for training model
plt.figure(figsize=(20,10))
plt.subplot(1,2,1)
plt.scatter(df['lstat'],df['medv'],marker='o',color='green')
plt.xlabel('lstat')
plt.ylabel('medv')
plt.subplot(1,2,2)
plt.scatter(df['rm'],df['medv'],marker='o',color='red')
plt.xlabel('rm')
plt.ylabel('medv')
plt.show()
```

```python
#creating dataframe for for train and test
new_df=df[['rm','lstat','medv']]
new_df
```

```python
from sklearn.model_selection import train_test_split as tts
from sklearn.linear_model import LinearRegression
xtrain,xtest,ytrain,ytest=tts(df[['rm','lstat']],df['medv'], test_size=0.30)
lr=LinearRegression()
lr.fit(xtrain,ytrain)
```

```python
lr.score(xtrain,ytrain)
```

#The model performance for training set is 0.667


```python
yh=lr.predict(xtest)
yh
```

```python
ytest
```

```python
comp=pd.DataFrame({'y':ytest,'yhat':yh})
comp
```

```python
from sklearn.metrics import r2_score
r2=r2_score(ytest,yh)
r2
```
